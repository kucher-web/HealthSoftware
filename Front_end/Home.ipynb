```python
# 导入函数库
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_excel("data.xlsx")
df_m = pd.read_excel("data_m.xlsx")
template_df = pd.read_excel("template.xlsx")
```

```python
# 提取第一列的所有值
values = template_df.iloc[:, 0].values.tolist()
# 筛选df和df_m中包含values的列名
filtered_columns1 = [col for col in df.columns if any(value in col for value in values)]
filtered_df1 = df[['星上时间']+filtered_columns1]
filtered_columns2 = [col for col in df_m.columns if any(value in col for value in values)]
filtered_df2 = df_m[['星上时间']+filtered_columns2]

# 将两个df合并成一个df
merged_df = pd.merge(filtered_df1, filtered_df2, on='星上时间', how='outer')
merged_df
```

```python
# 先把filtered_df转换为dataframe
result = merged_df
# 保存筛选结果到csv文件里面
result.to_excel('result.xlsx', index=False)
```
